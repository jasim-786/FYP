```python
# Path to the extracted dataset folder

dataset = '/content/Wheat'
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Input, GlobalAveragePooling2D, Dense, Concatenate
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3, MobileNet
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Directory for the dataset
data_dir = dataset  # Replace with actual dataset path

# Input layer
input_tensor = Input(shape=(224, 224, 3))

# Load both base models without top layers and without pre-trained weights
inception_base = InceptionV3(weights=None, include_top=False, input_tensor=input_tensor)
mobilenet_base = MobileNet(weights=None, include_top=False, input_tensor=input_tensor)

# Make both models trainable
inception_base.trainable = True
mobilenet_base.trainable = True

# Extract features
inception_features = GlobalAveragePooling2D()(inception_base.output)
mobilenet_features = GlobalAveragePooling2D()(mobilenet_base.output)

# Concatenate features
merged_features = Concatenate()([inception_features, mobilenet_features])
x = Dense(1024, activation='relu')(merged_features)
output = Dense(10, activation='softmax')(x)  # Update class count if needed

# Define hybrid model
hybrid_model = Model(inputs=input_tensor, outputs=output)

# Summary
hybrid_model.summary()

# Compile the model
hybrid_model.compile(optimizer=Adam(learning_rate=0.0001),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])

# ImageDataGenerators
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    os.path.join(data_dir, 'train'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    os.path.join(data_dir, 'validation'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    os.path.join(data_dir, 'test'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=False
)

# Train the model
history = hybrid_model.fit(train_generator, epochs=40, validation_data=val_generator)

# Plot training results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluation
def evaluate_model(test_generator, model, classes, title="Performance"):
    y_pred = model.predict(test_generator)
    y_pred = np.argmax(y_pred, axis=1)

    precision = precision_score(test_generator.labels, y_pred, average='weighted')
    recall = recall_score(test_generator.labels, y_pred, average='weighted')
    f1 = f1_score(test_generator.labels, y_pred, average='weighted')
    accuracy = accuracy_score(test_generator.labels, y_pred)

    print(f"\n{title}")
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print("Accuracy:", accuracy)

    cm = confusion_matrix(test_generator.labels, y_pred)
    cm_df = pd.DataFrame(cm, index=classes, columns=classes)
    print("Confusion Matrix:\n", cm_df)

# Call evaluation
evaluate_model(test_generator, hybrid_model, ['Brown_Rust', 'Healthy', 'Yellow_Rust'], title="Hybrid Model Performance")
```
